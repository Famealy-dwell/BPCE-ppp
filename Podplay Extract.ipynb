```python
import pandas as pd
import glob
```

```python
# Chemin vers le dossier contenant les CSV
path = "./"   # ou ex: "./data/"

# Liste de tous les CSV
all_files = glob.glob(path + "*.csv")

# Lecture et concat√©nation
df = pd.concat(
    (pd.read_csv(f) for f in all_files),
    ignore_index=True
)

# Aper√ßu
df
#df.to_excel("test.xlsx", index=False)
```

```python
df["Date (UTC)"] = pd.to_datetime(df["Date (UTC)"])
#df = df[(df["Cash Amount"] == 0) | (df["Cash Amount"].isna())]
```

```python
monthly_gross = (
    df
    .groupby(df["Date (UTC)"].dt.to_period("M"))["Gross"]
    .sum()
    .reset_index()
)

# Remettre la date au format timestamp (optionnel)
monthly_gross["Date (UTC)"] = monthly_gross["Date (UTC)"].dt.to_timestamp()
monthly_gross
```
